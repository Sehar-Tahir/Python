### To check the Normality of Data

# 1.Shapiro Wilk Test
Test whether a data sample has a Gaussian distribution(data is normal or not)

#### Assumptions
1. Observation in each sample are independent and identically distributed (iid)
2. Interpretation

* H0: the sample has Gaussian/Normal Distribution (normal data)
* H1: the sample has not Gaussian/Normal Distribution (not normal data)

## Steps:
* Import the shapiro function from scipy.stats
* Define your data as a list or array
* Call the shapiro function with your data to get the test statistic and p-value
* Print the results
* Conclude based on the p-value whether to reject or fail to reject the null hypothesis (H0)


```python
#Example of the Shapiro-Wilk test for normality

from scipy.stats import shapiro

data = [4.0, 3.2, 5.1, 6.3, 7.2, 5.9, 4.4, 5.6, 6.1, 5.0]
stat, p = shapiro(data)

print("stat=%.3f, p=%.3f" % (stat, p))

if p > 0.05:
	print("Gaussian/Normal Distribution - (Data is normal)")
else:
	print("Not Gaussian/Not Normal Distribution - (Data is not normal)")
```

# 2.Correlation Test
## 1- Pearson's Correlation Cofficient
Test whether two samples have a linear relationshp
### Assumptions
1. Observations in each sample are independent and identically distributed
2. Observations in each sample are normally distributed
3. Observations in each sample have the same varience
4. Interpretation

* H0: the two samples are independent.
* H1: there is a dependency btw the samples.

```python
# Example of Pearson's Correlation Test
from scipy.stats import pearsonr
# define data
data1 = [0.876, 8.654, 9.123, 5.234, 6.789, 7.890, 3.456, 2.345, 4.567]
data2 = [1.234, 7.890, 8.765, 4.321, 5.678, 6.789, 2.345, 3.456, 4.567]
# calculate Pearson's correlation
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably independent (No Correlation - accept H0)')
else:
    print('Probably dependent (Correlation Exsist - reject H0)')
```

## 2- Spearman's Rank Correlation
Test whether two samples have monotonic relationshp
### Assumptions
1. Observations in each sample are independent and identically distributed
2. Observations in each sample can b e ranked
3. Interpretation

* H0: the two samples are independent.
* H1: there is a dependency btw the samples.

```python
# Example of Spearman's Rank Correlation Test
from scipy.stats import spearmanr
# define data
data1 = [0.876, 8.654, 9.123, 5.234, 6.789, 7.890, -3.456, -2.345, 4.567]
data2 = [-1.234, 7.890, 8.765, 4.321, 5.678, 6.789, 2.345, -3.456, 4.567]
# calculate Spearman's correlation
stat, p = spearmanr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably independent (No Correlation - accept H0)')
else:
    print('Probably dependent (Correlation Exsist - reject H0)')
```

## 3- Chi-Squared Test
Test whether two categorical variables are related or independent
### Assumptions
1. Observations used in the calculation  of the contingency table are independent
2. 25 or more examples in each cell of the contigency table
3. Interpretation

* H0: the two samples are independent.
* H1: there is a dependency btw the samples.

```python
# Example of the Chi-Squared Test for Independence
from scipy.stats import chi2_contingency
# define the contingency table
table = [[10, 20, 30], [6,  9, 17]]
# calculate the Chi-Squared test
stat, p, dof, expected = chi2_contingency(table)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably independent (No Association - accept H0)')
else:
    print('Probably dependent (Association Exists - reject H0)')
```

# 2.Parametric Statistical Hypothesis Test
## 1- Student's t-test (independent)
Test whether the mean of two independent samples are significantly different
### Assumptions
1. Observations in each sample are independent and identically distributed
2. Observations in each sample are normally distributed
3. Observations in each sample have the same varience
4. Interpretation

* H0: the means of the samples are equal.
* H1: the means of the samples are not equal.

```python
# Example of the student's t-Test
from scipy.stats import ttest_ind
# define two sets of data
data1 = [2.9, 3.0, 2.5, 2.6, 3.2]
data2 = [3.8, 2.7, 4.0, 2.4, 2.8]
# calculate the t-test
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably the same distribution - same dataset (accept H0)')
else:
    print('Probably different distributions - different datasets (reject H0)')
```

```python
import seaborn as sns
#load dataset
ks = sns.load_dataset("titanic")

# Example of the student's t-Test
from scipy.stats import ttest_ind
# define two sets of data
data1 = ks['age'].dropna()
data2 = ks['fare'].dropna()
# calculate the t-test
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably the same distribution - same dataset (accept H0)')
else:
    print('Probably different distributions - different datasets (reject H0)')
```

## 2- Paired Student's t-test (dependent)
Test whether the mean of two independent samples are significantly different
### Assumptions
1. Observations in each sample are independent and identically distributed
2. Observations in each sample are normally distributed
3. Observations in each sample have the same varience
4. Observation across each sample are paired
5. Interpretation

* H0: the means of the samples are equal.
* H1: the means of the samples are not equal.

#

```python
# Example of the paired student's t-Test
from scipy.stats import ttest_rel
# define two sets of data
data1 = [2.9, 3.0, 2.5, 2.6, 3.2]
data2 = [3.8, 2.7, 4.0, 2.4, 2.8]
# calculate the t-test
stat, p = ttest_rel(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably the same distribution - same dataset (accept H0)')
else:
    print('Probably different distributions - different datasets (reject H0)')
```

## 3- Analysis of varience test (ANOVA)
Test whether the mean of two or more  independent samples are significantly different
### Assumptions
1. Observations in each sample are independent and identically distributed
2. Observations in each sample are normally distributed
3. Observations in each sample have the same varience
4. Interpretation

* H0: the means of the samples are equal.
* H1: one or more means of the samples are not equal.

```python
# Example of the analysis of variance (ANOVA) test
from scipy.stats import f_oneway
# define three sets of data
data1 = [2.9, 3.0, 2.5, 2.6, 3.2]
data2 = [3.8, 2.7, 4.0, 2.4, 2.8]
data3 = [2.8, 3.4, 3.7, 2.0, 3.0]
# calculate the ANOVA  test
stat, p = f_oneway(data1, data2, data3)
print('stat=%.3f, p=%.3f' % (stat, p))
# interpret the result
if p > 0.05:
    print('Probably the same distribution - same dataset (accept H0)')
else:
    print('Probably different distributions - different datasets (reject H0)')
```
