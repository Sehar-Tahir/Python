# 2- Multiple Regression

```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# import your data
df = pd.read_csv('SalaryMulti.csv')
df.head()
```

```python
# split the data into independent and dependent variables
x = df[['Total Experience', 'Team Lead Experience', 'Project Manager Experience', 'Certifications']]
y = df['Salary']
```

```python
# create and fit your model 
model = LinearRegression().fit(x, y)
model
```

```python
# view the model coefficients
model.coef_
```

```python
# view the model intercept
model.intercept_
```

```python
# make a prediction
model.predict([[2, 1, 0, 3]])  # 2 years total experience, 1 year team lead experience, 0 years project manager experience, 3 certifications
```

```python
# view the score (R^2 value) for the model
model.score(x, y)
```

```python
# train test split and view scores


# import library and split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
# create and fit your model
model = LinearRegression().fit(x_train, y_train)
print("Test Score: ", model.score(x_test, y_test))
print("Train Score: ", model.score(x_train, y_train))
```

```python
# plot multiple regression model
import matplotlib.pyplot as plt

# 3D plot for two independent variables against the dependent variable
fig = plt.figure(figsize=(10, 7))
myplot = fig.add_subplot(111, projection='3d')
myplot.scatter(df['Total Experience'], df['Team Lead Experience'], df['Salary'], c='r', marker='o')
myplot.set_xlabel('Total Experience')  
myplot.set_ylabel('Team Lead Experience')
myplot.set_zlabel('Salary')         
myplot.set_title('3D Scatter plot of Salary vs Total Experience and Team Lead Experience')
plt.show()
```
