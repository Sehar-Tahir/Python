# a- Decission Tree Classification

```python
#import libraries
import pandas as pd
df = pd.read_csv('BreastCancer.csv')
df.head()
```

```python
# data preprocessing - converting categorical data to numerical data
df['Estrogen Status'] = df['Estrogen Status'].replace({'Positive':1, 'Negative':0})
df['Progesterone Status'] = df['Progesterone Status'].replace({'Positive':1, 'Negative':0})
df.head()
```

```python
df.tail()
```

```python
# selection of input and output variables
x = df[['Age', 'Tumor Size', 'Estrogen Status', 'Progesterone Status']]
y = df['Status']
```

```python
x.head()
```

```python
y.head()
```

```python
#machine learning algorithm
from sklearn.tree import DecisionTreeClassifier

#create and fit the model
model = DecisionTreeClassifier().fit(x, y)

# make predictions
model.predict([[45, 30, 1, 0]]) # Example input: Age=45, Tumor Size=30, Estrogen Status=Positive, Progesterone Status=Negative
```

```python
#measure the accuracy of the model

#split the data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

model = DecisionTreeClassifier().fit(x_train, y_train)

predicted_values = model.predict(x_test)
predicted_values
```

```python
# check accuracy score
score = accuracy_score(y_test, predicted_values)
score
```

# Plots Display using Decision Tree Algorithm

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure()
model = DecisionTreeClassifier().fit(x, y)
plot_tree(model, filled=True)
plt.title("Decision Tree for Breast Cancer Status Prediction")
plt.show()
```

```python
#load sample dataset
import pandas as pd
import numpy as np
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
import matplotlib.pyplot as plt
import sklearn.tree as DecisionTreeClassifier
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

model = DecisionTreeClassifier().fit(x, y)
plot_tree(model, filled=True)
plt.title("Decision Tree for Iris Species Prediction")
plt.show()
```
