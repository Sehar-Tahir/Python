* The above data is really in wide range and we need to normalize and hard to compare
* Normalization change the values to the range of 0-1 (now both variable has similar influence on our models)

### Methods of Normalization

1.  Simple feature scaling
   * X(new)= x(old)/x(max)
2.  Min-Max method
3.  Z-score (standard score) -3 to +3
4.  Log transformation

```python
#simple feature scaling - normalization (0-1)
ks3 = ks1[['age in days', 'fare']].copy()                   #create a copy of the dataframe to avoid changing original dataframe
ks3['fare'] = ks3['fare']/ks3['fare'].max()
ks3['age in days'] = ks3['age in days']/ks3['age in days'].max()
ks3.head()
```

```python
# Min - Max  Method - Normalization - (X - Xmin) / (Xmax - Xmin) - better than above method
ks3['fare'] = (ks3['fare'] - ks3['fare'].min()) / (ks3['fare'].max() - ks3['fare'].min())
ks3.head()
```

```python
# Z-score method (standard score) - (value - mean) / std.deviation - most used method - normal distribution
ks3['fare'] = (ks3['fare'] - ks3['fare'].mean()) / ks3['fare'].std()
ks3.head()
```

```python
# log transformation
ks3['fare'] = np.log(ks3['fare'])  #log(0) is undefined so adding 1 to avoid it
ks3.head()
```

# Binning
* Grouping of values into smaller number of values (bins)
* Convert numeric into categories (child, bachay, boorhay) or 1-16, 17-30
* To have better understanding of groups
  - low vs mid vs high price

```python
bins = np.linspace(min(ks1['age in days']), max(ks1['age in days']), 4) #3 bins
age_groups = ['bachy', 'jawan', 'boorhay']   #labels for bins
ks1['age'] = pd.cut(ks1['age in days'], bins, labels=age_groups, include_lowest=True) #include_lowest means include the lowest value in the first bin
ks1[['age']]
```
